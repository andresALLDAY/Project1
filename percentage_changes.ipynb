```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from config import akey
```

```python
file = 'api-data/yearly_summary/prices_for_2016.csv'
stocks2016 = pd.read_csv(file)
stocks2016.sort_values(by=['Date'], inplace=True)
stocks2016.head(5)

```

```python
stocks2016.head()
stocks2016.drop(columns=['Open','High','Low','Daily Volume'], inplace=True)
stocks2016.head()
```

```python
pct_chg_2016 = stocks2016.groupby('Ticker',sort=False).apply(lambda x: (x['Close'].values[-1] - x['Close'].values[0]) / x['Close'].values[-1] * 100)\
    .reset_index(name='pct change')
pct_chg_2016.head()
```
